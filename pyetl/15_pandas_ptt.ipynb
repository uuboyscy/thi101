```python
import glob
import pandas as pd
```

```python
file_path_list = glob.glob(
    r"/Users/uuboy.scy/PycharmProjects/thi101/pyetl/res_gossiping/*.txt"
)
```

```python
data = []
for file_path in file_path_list:
    with open(file_path, "r", encoding="utf-8") as f:
        tmp_data = f.read().split("---split---")[1].split("\n")[1:-1]
        # print(tmp_data)
        data.append(tmp_data)
print(data)
```

```python
columns = [
    "推",
    "噓",
    "分數",
    "作者",
    "標題",
    "時間",
]
```

```python
df = pd.DataFrame(data=data, columns=columns)
```
